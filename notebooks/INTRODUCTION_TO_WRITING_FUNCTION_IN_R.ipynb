####**5. Utilisation de ...**

+ L’argument ellipse, ou point-point-point, vous permet de simplifier un peu votre code.

+ L’utilisation de point-point-point ici présente deux avantages, l’un évident et l’autre moins. 

+ L’avantage évident est que c’est moins de frappe pour vous. 

+ Le plus subtil, qui est plus important, c’est que si la signature de la fonction interne change, votre code n’a pas besoin de changer pour y correspondre. 

+ Par exemple, si la fonction mean changeait le nom de son argument na-dot-rm en quelque chose d’autre comme remove_missing, votre code fonctionnerait automatiquement toujours. 

+ L’inconvénient est que ce n’est peut-être pas ce que vous voulez. 

+ Étant donné qu’un changement de nom d’argument pourrait casser le code écrit par vos utilisateurs, vous préféreriez peut-être que le changement casse simplement votre code afin que vous puissiez décider comment le gérer. 

+ Dans l’ensemble, qu’il soit préférable d’utiliser point-point-point ici ou non dépend de la confiance que vous accordez à la fonction interne et de la stabilité dont vous avez besoin pour votre signature. 

+ Enfin, point-point-point n’est pas aussi clair pour les utilisateurs, car ils devraient lire la documentation de la fonction interne pour savoir quoi taper.

```r
is_leap_year <- function(year) {
  # If year is div. by 400 return TRUE
  if(year %% 400 == 0) {
    return(TRUE)
  }
  # If year is div. by 100 return FALSE
  if(year %% 100 == 0) {
    return(FALSE)
  }  
  # If year is div. by 4 return TRUE
  if (year %% 4 == 0){
    TRUE
  }
  
  
  # Otherwise return FALSE
  else {
  FALSE
  }
}
```

+ $X$ est recherché dans l’environnement mondial, mais il n’est pas là. 

+ L’environnement de la fonction n’est pas un parent de l’environnement global, c’est plutôt un enfant. 

+ Donc, R ne peut pas regarder là-bas, ce qui signifie que $x$ ne peut pas être trouvé, et vous obtenez une erreur. 

+ Ainsi, une fonction peut rechercher des variables en dehors de son environnement, mais l’inverse n’est pas vrai : vous ne pouvez pas regarder à l’intérieur de l’environnement de la fonction de l’extérieur.

```r
# Write a function to convert lb to kg
lbs_to_kgs <- function(lbs) {lbs * 0.45359237}

# Write a function to convert bushels to lbs
bushels_to_lbs <- function(bushels, crop) {
  # Define a lookup table of scale factors
  c(barley = 48, corn = 56, wheat = 60) %>%
    # Extract the value for the crop
    extract(crop) %>%
    # Multiply by the no. of bushels
    multiply_by(bushels)
}

# Write a function to convert bushels to kg
bushels_to_kgs <- function(bushels, crop) {
  bushels %>%
    # Convert bushels to lbs for this crop
    bushels_to_lbs(crop) %>%
    # Convert lbs to kgs
    lbs_to_kgs()
}

# Write a function to convert bushels/acre to kg/ha
bushels_per_acre_to_kgs_per_hectare <- function(bushels_per_acre, crop = c("barley", "corn", "wheat")) {
  # Match the crop argument
  crop <- match.arg(crop)
  bushels_per_acre %>%
    # Convert bushels to kgs for this crop
    bushels_to_kgs(crop) %>%
    # Convert harmonic acres to ha
    harmonic_acres_to_hectares()
}
```

####**3. Modèles linéaires vs modèles additifs généralisés**
+ Voici un modèle de code pour exécuter un modèle de régression linéaire. 

+ Vous appelez lm, en passant une formule avec la réponse à gauche et les variables explicatives à droite, puis définissez l’argument de données sur le bloc de données contenant les données.
+ Pour exécuter un GAM, vous devez d’abord charger le package mgcv, qui signifie « véhicule de calcul GAM mixte ». 
+ Après cela, le code n’a que deux différences par rapport au cas linéaire.
+ Plutôt que d’appeler lm, vous appelez gam.
+ L’autre différence est que les variables explicatives de la formule peuvent être enveloppées dans s. 
+La fonction s signifie « rendre l’effet de cette variable lisse ». Très grossièrement, cela signifie que vous avez une réponse non linéaire, mais plutôt que de viser un ajustement parfait aux données, vous voulez quelque chose qui change progressivement, donnant une ligne lisse plutôt qu’une ligne dentelée.

####**4. Prédire les GAM**
+ Pour créer un bloc de données de prédictions de modèle, il y a trois étapes. 
+ Tout d’abord, vous devez spécifier un bloc de données des cas que vous souhaitez prédire. 
+ Autrement dit, chaque colonne de la trame de données doit contenir les valeurs d’une variable explicative dans le modèle. 
+ Deuxièmement, vous appelez predict, en passant le modèle GAM et le cadre de données des cas à prédire. 
+ Vous devez également spécifier que le type de prédiction est la réponse. 
+ Les réponses prédites sont données sous forme de vecteur, mais pour plus de commodité, il est utile de les stocker sous forme de colonne dans le cadre de données des cas à prédire. Vous pouvez muter ce bloc de données pour ajouter la colonne.

```r
# Wrap the model code into a function
run_gam_yield_vs_year_by_region <- function(data) {
  gam(yield_kg_per_ha ~ s(year) + census_regions, data = data)
}

# Try it on the wheat dataset
run_gam_yield_vs_year_by_region(wheat)
```

```r
corn_model <- run_gam_yield_vs_year_by_region(corn)
wheat_model <- run_gam_yield_vs_year_by_region(wheat)
```

```r
# Make predictions in 2050  
predict_this <- data.frame(
  year = 2050,
  census_region = census_regions
) 

# Predict the yield
pred_yield_kg_per_ha <- predict(corn_model, predict_this, type = "response")

predict_this %>%
  # Add the prediction as a column of predict_this 
  mutate(pred_yield_kg_per_ha = pred_yield_kg_per_ha)
```

```r
# Wrap this prediction code into a function
predict_yields <- function(model, year) {
  predict_this <- data.frame(
    year = year,
    census_region = census_regions
  )
  pred_yield_kg_per_ha <- predict(model, predict_this, type = "response")
  predict_this %>%
    mutate(pred_yield_kg_per_ha = pred_yield_kg_per_ha)
}

# Try it on the wheat dataset
predict_yields(wheat_model, year = 2050)
```
