###**Explorer les données numériques**

####**1. Explorer les données numériques**

+ Dans ce chapitre, nous élargirons notre boîte à outils de techniques exploratoires pour englober les données numériques. 

+ Les données numériques sont des données qui prennent la forme d’un nombre, mais où ces nombres représentent en fait une valeur sur la ligne numérique.
+ L’ensemble de données avec lequel nous allons travailler contient des informations sur les voitures qui étaient en vente aux États-Unis au cours d’une certaine année.

####**2. Ensemble de données sur les voitures**

+ Nous pouvons utiliser la fonction de structure, s-t-r, pour en savoir plus sur chacune des variables. Nous apprenons que nous avons 428 observations, ou cas, et 19 variables. 

+ Contrairement à la plupart des affichages de données, la fonction structure place chacune des variables sous forme de ligne, avec son nom suivi de son type de données, suivi des premières valeurs. 

+ Les noms de voiture sont des chaînes de caractères, qui sont comme des facteurs, sauf qu’il est courant que chaque cas prenne une valeur unique. L-o-g-i, c’est pour les variables logiques, un autre cas simple d’une variable catégorielle où il n’y a que deux niveaux. 

+ Par exemple, chaque voiture prendra VRAI ou FAUX selon qu’il s’agit d’une voiture de sport. 

+ Nous pouvons voir que le dernier ensemble de variables est soit i-n-t pour entier, soit n-u-m pour numérique. 

+ Ce sont en fait deux variables numériques, mais les entiers sont discrets et le numérique est continu. 

+ Si vous regardez ncyl, c’est le nombre de cylindres, il est répertorié comme un entier, mais il n’y a que quelques valeurs différentes qu’il peut prendre, donc il se comporte un peu comme une variable catégorielle. 
+ Construisons quelques graphiques pour nous aider à explorer ces données.

####**3. Dotplot**
+ La façon la plus directe de représenter des données numériques est d’utiliser un diagramme à points, où chaque cas est un point placé à sa valeur appropriée sur l’axe des $x$, puis empilé lorsque d’autres cas prennent des valeurs similaires. 

+ Il s’agit d’une forme de graphique où il n’y a aucune perte d’information; Vous pourriez en fait reconstruire parfaitement l’ensemble de données si on vous donnait ce graphique. 

+ Comme vous pouvez l’imaginer, cependant, ces intrigues commencent à devenir difficiles à lire à mesure que le nombre de cas devient très important.

####**4. Histogramme**
+ L’un des graphiques les plus courants à utiliser est un histogramme, qui résout ce problème en agrégeant les points dans des bacs sur l’axe des $x$, puis en mappant la hauteur de la barre au nombre de cas qui tombent dans ce bac. 

+ En raison du binning, il n’est pas possible de reconstruire parfaitement l’ensemble de données : ce que nous gagnons, c’est une image plus large de la forme de la distribution. 

+ Si la nature progressive de l’histogramme vous irrite, alors vous aimerez le diagramme de densité.

####**5. Diagramme de densité**
+ Le diagramme de densité représente la forme de l’histogramme à l’aide d’une ligne lisse. 

+ Cela fournit une représentation toujours plus grande de la forme de la distribution, vous ne voudrez donc l’utiliser que lorsque vous avez un grand nombre de cas.

###**Mesures de la variabilité**

####**1. Mesures de la variabilité**

+ Il en résulte 11 différences, certaines positives, d’autres négatives. 

+ Nous aimerions réduire toutes ces différences à une seule mesure de variabilité, alors additionnons-les. 

+ C’est la notation scientifique de R, disant que la somme est -1 (point) 42 fois 10 à -14. 

+ Ce nombre est essentiellement nul. 

+ De toute évidence, quelque chose s’est mal passé parce que nous pouvons dire qu’il y a une variabilité dans cet ensemble de données, mais notre mesure ne l’a pas détectée. 

+ Le problème est que les points positifs et négatifs se sont annulés. 

+ C’est facile à corriger: vous pouvez résoudre au carré chaque différence pour vous débarrasser des points négatifs. 

+ Cette nouvelle mesure est meilleure, mais elle a une propriété indésirable: elle ne fera que croître au fur et à mesure que vous ajouterez de données.

OK, maintenant cela semble être une mesure utile: vous trouvez le centre des données, puis trouvez la distance au carré entre les observations et cette moyenne moyenne sur l’ensemble de l’ensemble de données. 

+ Si vous changez le n en n-1, vous vous retrouvez avec ce qu’on appelle la variance de l’échantillon, l’une des mesures les plus utiles de la propagation d’une distribution. 

+ Dans R, cette statistique est encapsulée dans la fonction v-a-r pour la variance.

####**4. Insérez le titre ici...**
+ Une autre mesure utile est la racine carrée de ce nombre, appelée écart type de l’échantillon ou simplement sd dans R. 

+ Ce qui est pratique à propos de l’écart-type de l’échantillon, c’est qu’une fois calculé, il se trouve dans les mêmes unités que les données originales. 

+ Dans ce cas, nous pouvons dire que l’écart-type de l’espérance de vie de ces 11 comtés est de 1 (point) 69 ans. 

+ En comparaison, la variance de cet échantillon est de 2 (points) 87 ans au carré, ce qui est une unité sur laquelle nous n’avons aucune intuition réelle. 

+ Il y a deux autres mesures de propagation qu’il est bon de connaître. 

+ La plage interquartile, ou IQR, est la distance entre les deux nombres qui coupent les 50 % intermédiaires de vos données. 

+ Cela devrait sembler familier à la discussion sur les diagrammes de boîte: 
  + **la hauteur de la boîte est exactement l’IQR**. 

+ Nous pouvons soit obtenir les premier et troisième quartiles de la fonction sommaire et prendre leur différence, soit utiliser la fonction IQR intégrée. 

+ La mesure finale est simplement la plage des données: la distance entre le maximum et le minimum. 

+ **max et min** sont en effet des fonctions dans R, mais vous pouvez aussi utiliser le diffrangex imbriqué. 

+ Pour n’importe quel ensemble de données, vous pouvez calculer ces quatre statistiques, mais lesquelles sont les plus significatives ? 

+ Le plus couramment utilisé dans la pratique est l’écart-type, c’est donc souvent un bon point de départ. 

+ Mais que se passe-t-il si l’ensemble de données contient des observations extrêmes?

+ Ces mesures sont sensibles aux valeurs extrêmes de la même manière que la moyenne est une mesure du centre. 

+ Si vous recalculez la plage, elle augmentera certainement car elle est complètement déterminée par les valeurs extrêmes. 

+ Pour cette raison, la plage n’est pas souvent utilisée. 

+ Si vous recalculez l’IQR, cependant, vous voyez qu’il n’a pas bougé. Parce que cette observation est toujours la plus élevée, les quartiles n’ont pas bougé. 

+ Cela révèle une bonne raison d’utiliser l’IQR: dans les situations où votre ensemble de données est fortement biaisé ou a des observations extrêmes.

###**Forme et transformations**

####**1. Forme et transformations**
+ Il y a généralement quatre caractéristiques des distributions qui sont intéressantes. 

+ Les deux premiers que nous avons déjà couverts : **le centre et la propagation ou la variabilité de la distribution**. 

+ La troisième est la forme de la distribution, qui peut être décrite en termes de modalité et d’inclinaison.

####**2. Modalité**
+ **La modalité d’une distribution est le nombre de bosses proéminentes qui apparaissent dans la distribution**. 

+ S’il y a un seul mode, comme dans une courbe en cloche, il est dit unimodal. 

+ S’il y a deux modes importants, on parle de bimodal. 

+ S’il a trois modes ou plus, la convention est de le qualifier de multimodal. 

+ Il y a un dernier cas : quand il n’y a pas de mode distinct. Étant donné que la distribution est plate sur toutes les valeurs, elle est appelée uniforme. 

+ L’autre aspect de la forme de la distribution concerne son inclinaison.

####**3. Inclinaison**
+ Si une distribution a une longue queue qui s’étend vers la droite, on parle de **« droit-asymétrique »**.

####**4. Inclinaison**
+ Si cette longue queue s’étend vers la gauche, on parle de « gauche ». 

+ Si vous avez du mal à vous rappeler lequel est lequel, rappelez-vous simplement que le biais est l’endroit où se trouve la longue traîne.

####**5. Inclinaison**
+ Si aucune queue n’est plus longue que l’autre, la distribution est dite « symétrique ».

####**6. Forme du revenu**

+ Comparons les distributions du revenu personnel médian au niveau des comtés sur la côte ouest et dans le reste du pays pour voir quelle forme prennent ces distributions.

+ Le graphique qui en résulte montre deux courbes, le bleu représentant la distribution de la côte Ouest et le rose représentant les comtés qui ne font pas partie de la côte Ouest.
