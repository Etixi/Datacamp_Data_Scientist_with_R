###**Le verbe de comptage**
####**1. Le verbe de comptage**
+ Jusqu’à présent dans ce cours, vous avez appris à sélectionner des variables dans un ensemble de données, à filtrer ou à trier les observations qui vous intéressent et à ajouter de nouvelles variables avec le verbe muter. 

+ Mais, jusqu’à présent, vous avez travaillé au même niveau que les données initiales, où chaque observation correspond à un comté américain. 

+ Dans ce deuxième chapitre, vous apprendrez à agréger les données: à prendre de nombreuses observations et à les résumer en une seule. 

+ Il s’agit d’une stratégie commune de science des données pour rendre les ensembles de données gérables et interprétables.

####**2. Dénombrement**

+ La façon simple d’agréger les données est de les compter : pour connaître le nombre d’observations. Le verbe dplyr pour cela est **count()**. 

+ Le résultat est une table à une ligne, avec une colonne appelée n. Cela nous indique qu’il y a 3 138 observations dans le tableau, et donc autant de comtés aux États-Unis.

```r
counties %>%  
  count()
```

####**3. Variable de comptage**

+ Compter les données totales est un peu utile, mais la valeur réelle du verbe est lorsque vous lui donnez une variable spécifique à compter. 

+ Par exemple, vous pouvez compter le nombre de comtés dans chaque État. Notez que le résultat comporte 50 observations : une pour chacun des 50 états. 

+ Nous avons agrégé plus de trois mille observations en un nombre plus gérable. La deuxième colonne, n, nous dit qu’il y a 67 comtés en Alabama, 28 en Alaska, et ainsi de suite.



```r
#Count variable
counties %>%  
  count(state)
```

###**4. Compter et trier**

+ Lorsque les scientifiques des données explorent un ensemble de données, nous sommes souvent intéressés par le tri des données comptées pour trouver les observations les plus courantes. 

+ Le verbe count prend un deuxième argument qui est très utile pour cela. 

+ Nous pouvons ajouter que le tri par virgule est égal à TRUE, et les lignes résultantes seront triées des observations les plus courantes aux moins courantes. 

+ Cela nous indique que le Texas est l’État qui compte le plus de comtés, suivi de la Géorgie et de la Virginie.


```r
#Count and sort
counties %>%  
  count(state, sort =TRUE)
```


####**5. Dénombrement de la population**

####**6. Ajoutez du poids**

+ Et si, au lieu de trouver le nombre de comtés dans chaque État, nous voulions connaître le nombre total de personnes? 

+ Nous pouvons ajouter l’argument wt, qui signifie « poids », est égal à la population. 

+ Cela signifie que la colonne n sera pondérée par la population. 

+ En conséquence, au lieu de voir le nombre de comtés dans chaque État, nous verrions la population totale. 

+ Ici, nous pouvons voir que la Californie est l’État américain avec la population la plus élevée, suivie du Texas et de New York.

####**8. Regrouper par**

+ Notez que le résultat est toujours regroupé par état : vous pouvez voir « Groupes: état » en haut. Lorsque vous utilisez summarize sur un tableau comportant plusieurs groupes, seul le dernier d’entre eux est « décollé ». 

+ Ceci est utile lorsque vous souhaitez continuer à effectuer des résumés ou des agrégations supplémentaires.

####**9. Dissocier**

```r
#Summarize
counties %>%  
  summarize(total_population =sum(population))
```

```r
#Aggregate and summarize
counties %>%  
  summarize(total_population =sum(population),            
  average_unemployment = mean(unemployment))
```

```r
# Aggregate with in groups
counties %>%  
  group_by(state)%>%  
    summarize(total_pop =sum(population),            
    average_unemployment = mean(unemployment))
```

```r
# Arrange
counties %>%  
  group_by(state)%>%  
    summarize(total_pop =sum(population),            
    average_unemployment = mean(unemployment))%>%  
      arrange(desc(average_unemployment))
```

```r
#Metro column
counties %>%  
  select(state, metro, county, population)
```

```r
#Groupby
counties %>%  
  group_by(state, metro)%>%  
    summarize(total_pop =sum(population), .groups = 'drop')
```

```r
#Ungroup
counties %>%  
  group_by(state, metro)%>%  
    summarize(total_pop =sum(population), .groups = 'drop')%>%  
      ungroup()
```
