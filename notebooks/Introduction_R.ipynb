####**Qu’est-ce qu’un facteur et pourquoi l’utiliseriez-vous?**

+ Dans ce chapitre, vous plongez dans le monde merveilleux des **facteurs**.

+ **Le terme facteur fait référence à un type de données statistiques utilisé pour stocker des variables catégorielles**.

+ **La différence entre une variable catégorielle et une variable continue est qu’une variable catégorielle peut appartenir à un nombre limité de catégories**.

+ Une variable continue, en revanche, peut correspondre à un nombre infini de valeurs.

+ Il est important que R sache s’il s’agit d’une variable continue ou catégorielle, car les modèles statistiques que vous développerez à l’avenir traitent les deux types différemment. (Vous verrez plus tard pourquoi c’est le cas.)

+ Un bon exemple de variable catégorielle est le sexe.

+ Dans de nombreuses circonstances, vous pouvez limiter les catégories de sexe à « Homme » ou « Femme ». (Parfois, vous pouvez avoir besoin de différentes catégories.

+ Par exemple, vous devrez peut-être tenir compte de la variation chromosomique, des animaux hermaphrodites ou de différentes normes culturelles, mais vous aurez toujours un nombre fini de catégories.)

####**Qu’est-ce qu’un facteur et pourquoi l’utiliseriez-vous? (3)**
+ Il existe deux types de variables catégorielles : **une variable catégorielle nominale et une variable catégorielle ordinale.**

+ **Une variable nominale est une variable catégorielle sans ordre implicite**.

+ Cela signifie qu’il est impossible de dire que « l’un vaut plus que l’autre ».

+ Par exemple, pensez à la variable catégorielle animals_vector avec les catégories "Elephant","Giraffe", "Donkey", "Horse".

+ Ici, il est impossible de dire que l’un se situe au-dessus ou au-dessous de l’autre. (Notez que certains d’entre vous pourraient ne pas être d’accord ;-) ).

+ **En revanche, les variables ordinales ont un ordre naturel**.

+ Considérons par exemple la variable catégorielle temperature_vector avec les catégories:"Low", "Medium", "High".

+ Ici, il est évident que "Medium" se trouve au-dessus de "Low", et "High" se tient au-dessus de "Medium".

####**Instructions**

+ Soumettez la réponse pour vérifier comment R construit et imprime les variables nominales et ordinales.

+ Ne vous inquiétez pas si vous ne comprenez pas encore tout le code, nous y reviendrons.




```r
# Animals
animals_vector <- c("Elephant", "Giraffe", "Donkey", "Horse")
factor_animals_vector <- factor(animals_vector)
factor_animals_vector

# Temperature
temperature_vector <- c("High", "Low", "High","Low", "Medium")
factor_temperature_vector <- factor(temperature_vector, order = TRUE, levels = c("Low", "Medium", "High"))
factor_temperature_vector
```

####**Classement: Sorting**

+ Faire et créer des classements est l’une des affaires préférées de l’humanité.

+ Ces classements peuvent être utiles (meilleures universités du monde), divertissants (stars de cinéma les plus influentes) ou inutiles (meilleur sosie 007).

+ Dans l’analyse des données, vous pouvez trier vos données en fonction d’une certaine variable dans le jeu de données.

+ Dans R, cela se fait à l’aide de la fonction **order()**.

+ **order()**: est une fonction qui vous donne la position classée de chaque élément lorsqu’il est appliqué sur une variable, comme un vecteur par exemple :

      a <- c(100, 10, 1000)
      order(a)
      [1] 2 1 3

+ 10, qui est le deuxième élément de a, est le plus petit élément, donc 2 vient en premier dans la sortie de order(a).

+ 100, qui est le premier élément de a, est le deuxième plus petit élément, donc 1 vient deuxième dans la sortie de order(a).

+ Cela signifie que nous pouvons utiliser la sortie de order(a) pour remanier a :

      a[order(a)]
      [1]   10  100 1000

+ Félicitations! À ce stade du cours, vous êtes déjà familier avec:

  > **Vecteurs (tableau unidimensionnel)**: peut contenir des valeurs numériques, de caractères ou logiques. Les éléments d’un vecteur ont tous le même type de données.

  > **Matrices (tableau bidimensionnel)**: peut contenir des valeurs numériques, de caractères ou logiques. **Les éléments d’une matrice ont tous le même type de données**.

  > **Trames de données ou Dataframe** (objets bidimensionnels) : peuvent contenir des valeurs numériques, de caractères ou logiques. **Au sein d’une colonne, tous les éléments ont le même type de données, mais différentes colonnes peuvent être de type de données différent**.

####**Listes, pourquoi en auriez-vous besoin? (2)**
+ Une liste dans R est similaire à votre liste de tâches au travail ou à l’école: les différents éléments de cette liste diffèrent probablement par la longueur, les caractéristiques et le type d’activité à effectuer.

+ Une liste dans R vous permet de rassembler une variété d’objets sous un seul nom (c’est-à-dire le nom de la liste) de manière ordonnée.

+ **Ces objets peuvent être des matrices, des vecteurs, des trames de données, même d’autres listes, etc**.

+ Il n’est même pas nécessaire que ces objets soient liés les uns aux autres de quelque manière que ce soit.

+ Vous pourriez dire qu’une liste est une sorte de type de super données: vous pouvez y stocker pratiquement n’importe quelle information!



####**Création d’une liste**
+ Laissez-nous créer notre première liste! Pour construire une liste, utilisez la fonction **list()**:

      my_list <- list(comp1, comp2 ...)

+ Les arguments de la fonction de liste sont les composants de liste.

+ N’oubliez pas que ces composants peuvent être des matrices, des vecteurs, d’autres listes, ...

####**Création d’une liste nommée**
+ Bravo, vous êtes sur une lancée!

+ Tout comme sur votre liste de tâches, vous voulez éviter de ne pas savoir ou de ne pas vous souvenir de ce que les composants de votre liste représentent.

+ C’est pourquoi vous devriez leur donner des noms:

      my_list <- list(name1 = your_comp1,
                      name2 = your_comp2)


+ Cela crée une liste avec les composants qui sont nommés, name1 et name2 ainsi de suite.

+ Si vous souhaitez nommer vos listes après les avoir créées, vous pouvez utiliser la fonction **names()** comme vous l’avez fait avec les vecteurs.

+ Les commandes suivantes sont entièrement équivalentes à l’affectation ci-dessus :

      my_list <- list(your_comp1, your_comp2)

      names(my_list) <- c("name1", "name2")

####**Sélection d’éléments dans une liste**
+ Votre liste sera souvent construite à partir de nombreux éléments et composants.

+ Une façon de sélectionner un composant consiste à utiliser la position numérotée de ce composant.

+ Par exemple, pour « saisir » le premier composant de shining_list vous tapez

      shining_list[[1]]

+ Un moyen rapide de vérifier cela est de le taper dans la console.

+ Important à retenir : pour sélectionner des éléments à partir de vecteurs, vous utilisez des crochets simples :[ ].

+ Ne les mélangez pas!

+ Vous pouvez également vous référer aux noms des composants, avec [[ ]] ou $ avec lesigne. Les deux sélectionneront le cadre de données représentant les avis :

      shining_list[["reviews"]]
      shining_list$reviews

+ Outre la sélection de composants, vous devez souvent sélectionner des éléments spécifiques parmi ces composants.

+ Par exemple, avec shining_list[[2]][1] vous sélectionnez dans actors le deuxième composant,(shining_list[[2]]), le premier élément ([1]).
