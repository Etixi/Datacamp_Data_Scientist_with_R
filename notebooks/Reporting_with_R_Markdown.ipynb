###**L’en-tête YAML**

####**1. L’en-tête YAML**
+ Jusqu’à présent dans ce chapitre, nous avons appris deux éléments d’un fichier Markdown, le code et le texte. 

+ Parlons maintenant de l’en-tête YAML.

####**2. En-tête YAML**
+ **L’en-tête YAML contient les métadonnées du fichier sous forme de paires clé-valeur qui correspondent aux informations sur le document**. 

+ Il apparaît en haut du fichier markdown et commence et se termine par deux jeux de trois traits d’union. 

+ Lorsque nous créons un nouveau document markdown, par défaut, il inclut le titre, l’auteur et le format de sortie. 

+ Le titre et l’auteur sont fournis sous forme de chaînes. 

+ L’en-tête YAML nécessite une mise en forme spécifique pour le rendu, et il est important d’utiliser des deux-points après chaque champ, d’utiliser des coches rétroactives ou des informations de liste sous forme de chaînes si nécessaire et d’éviter d’ajouter des espaces supplémentaires. 

+ Après avoir modifié l’un des champs de l’en-tête YAML, nous tricoterons à nouveau le document pour voir les modifications reflétées dans le rapport final.

####**3. Types de sorties**

+ Le type de sortie du fichier Markdown est également spécifié dans l’en-tête YAML. 

+ Tout au long du cours, nous utiliserons deux types de sortie différents, **les fichiers HTML et PDF**. 

+ Nous pouvons basculer manuellement entre ces types en modifiant la section de sortie de l’en-tête YAML pour indiquer **html_document ou pdf_document**.

####**4. Exercices de cours**
+ Dans les exercices tout au long du cours, nous pouvons utiliser le menu déroulant à côté du bouton « Tricoter » pour basculer entre ces types de sortie.

####**5. Ajout manuel de la date**
+ Une autre information que nous pouvons ajouter à l’en-tête YAML est la date. 

+ Nous pouvons le saisir manuellement en ajoutant la date sous forme de chaîne.

####**6. Ajout automatique de la date**
+ Cependant, la saisie manuelle de la date n’est pas vraiment évolutive. 

+ Au lieu de cela, nous pouvons souhaiter que la date soit ajoutée automatiquement. 

+ Pour ce faire, ajoutez **la lettre r, un espace et la fonction Sys-dot-Date()** dans les backticks sous forme de chaîne. 

+ Lorsque nous tricotons le document, nous voyons que la date du jour est ajoutée au rapport avec le format année-mois-jour.

####**7. Mise en forme de la date**
+ Nous souhaiterons peut-être modifier davantage la date afin que la date soit formatée différemment dans le rapport. 

+ Nous pouvons utiliser la syntaxe suivante pour y parvenir: **format r et, entre parenthèses, Sys-dot-time() suivi du format que nous voulons que la date ait**. 

+ Ici, nous incluons le jour numérique, le nom complet du mois et l’année à quatre chiffres, en utilisant le pourcentage minuscule d % majuscule B % capital Y.

####**8. Options de formatage de la date**


+ Il existe un certain nombre d’options pour modifier le format de la date. 

+ Certaines options de texte pour la date incluent le pourcentage majuscule A ou le pourcentage minuscule a pour le jour de la semaine, et le pourcentage majuscule B ou le pourcentage minuscule b pour le mois; La lettre majuscule spécifie le nom complet, tandis que la lettre minuscule spécifie le nom abrégé. 

+ Certaines options numériques pour la date incluent **le pourcentage minuscule d pour le jour décimal, le pourcentage minuscule m pour le mois décimal et le pourcentage majuscule Y ou le pourcentage minuscule y pour l’année**. 

+ Comme précédemment, la lettre majuscule Y spécifie l’année complète à quatre chiffres, tandis que la lettre minuscule y spécifie l’année à deux chiffres.

    + https://www.rdocumentation.org/packages/base/topics/strptime

####**9. Exemple de formatage de date**


+ Nous pouvons également ajouter des virgules, des traits d’union ou des barres obliques inverses que nous voulons inclure dans la date. 

+ Dans cet exemple, nous avons modifié la date afin qu’elle s’affiche sous la forme d’une virgule mois jour année.

####**10. Autres options de date**
+ Au-delà de la modification du format de la date, nous pouvons inclure du texte qui apparaît à côté de la date.

###**Options de tracé**

####**1. Options de tracé**
+ Maintenant que nous avons ajouté des graphiques au rapport, voyons comment modifier chaque tracé et comment ils apparaissent tout au long du rapport.

####**2. Dimensions de la figure**
+ Souvent, nous voudrons modifier les dimensions de la figure. 

+ Il y a plusieurs façons de le faire. La première consiste à utiliser fig-dot-width et fig-dot-height pour spécifier la largeur et la hauteur de la figure en pouces dans le rapport final.

####**3. Dimensions de la figure**

+ Par exemple, si nous voulons que le chiffre soit de 5 pouces sur 3, nous mettons une virgule après le nom du morceau de code dans les accolades et incluons figue-point-largeur égale 5 et figue-point-hauteur égale 3. 

+ Notez que chaque option supplémentaire est séparée par une virgule.

####**4. Dimensions de la figure**
+ Alternativement, au lieu de spécifier ces dimensions séparément, nous pouvons utiliser fig-dot-dim égal à c 5 virgule 3 entre parenthèses, et nous obtiendrons le même résultat avec un seul argument au lieu de deux.

####**5. Dimensions de sortie**
+ Nous sommes également en mesure de modifier les dimensions de la figure en utilisant la largeur du point et la hauteur du point de sortie.

####**6. Dimensions de sortie**
+ Ces arguments déterminent les dimensions de la sortie des images dans le document tricoté et sont spécifiés sous forme de pourcentages. 

+ Par exemple, la largeur des points est égale à 50 %, ce qui aura pour conséquence que la figure occupera 50 % de la largeur du rapport une fois qu’il est tricoté. 

+ Nous incluons le pourcentage dans les citations. 

+ Ici, nous utilisons des guillemets simples, mais des guillemets doubles peuvent également être utilisés.

+ L’utilisation spécifique de guillemets simples ou doubles est importante lorsque des guillemets sont utilisés entre guillemets, par exemple, lorsque nous avons ajouté une mise en forme au champ de date dans l’en-tête YAML.

####**7. Alignement des figures**
+ Nous pouvons également modifier l’alignement de chaque figure dans le rapport à l’aide de fig-dot-align. 

+ Les options pour cela sont gauche, droite ou centre, répertoriées entre guillemets.

####**8. Alignement des figures**
+ Ici, nous avons utilisé l’alignement à gauche pour la figure et nous voyons que cela se reflète dans le document tricoté.

####**9. Options locales et globales**
+ Lorsque nous définissons les options pour les figures de chaque bloc de code, nous pouvons remarquer que certaines options de bloc de code sont répétées tout au long du rapport. 

+ Lorsque cela se produit, nous pouvons envisager de définir ces options globalement plutôt que localement. 

+ Lorsque nous spécifions des options localement, nous les ajoutons à chaque bloc de code et elles n’affectent que le code contenu dans ce segment. 

+ Si nous définissons les options globalement à la place, elles s’appliqueront à toutes les figures des blocs de code du rapport.

####**10. Définition globale des options**
+ Notez le segment de code qui apparaît lorsque nous créons un nouveau fichier Markdown, au début du fichier, dans le bloc de code nommé setup. 

+ Dans ce morceau, le paquet de tricot est appelé avec tricot et deux deux-points. 

+ Le package knitr est ce qui exécute chaque morceau de code et tricote le document. Dans ce segment, knitr utilisera les options spécifiées comme valeurs globales par défaut pour chaque segment du document lorsque le fichier est tricoté.

####**6. Ajout de tables avec kable()**

###**Options de bloc de code**

####**1. Options de bloc de code**
+ Nous avons appris à ajouter, étiqueter et modifier des morceaux de code. 

+ Discutons maintenant de certaines options de bloc de code que nous pouvons utiliser pour modifier le rapport de tricot.

####**2. Le bloc de code de données**
+ Regardons le morceau de code que nous avons vu dans les exercices tout au long du cours, appelé données. Le bloc de données a chargé les données et les packages nécessaires pour chaque exercice.

####**3. Le bloc de code de données**
+ Notez que, après le nom du morceau de code, nous voyons inclure est égal à FALSE à l’intérieur des accolades.

####**4. L’option include**
+ L’option include détermine si le code et les résultats apparaissent ou non dans le rapport. Lorsque include a la valeur FALSE, le code et les résultats n’apparaissent pas dans le rapport. 

+ Lorsque le fichier est tricoté, le bloc de code s’exécute et les résultats peuvent être utilisés dans tout le fichier, comme le bloc de données dans chaque exercice. 

+ L’argument par défaut pour include est TRUE. 

+ Étant donné que l’inclusion n’est spécifiée dans aucun autre bloc de code dans les exercices précédents, le code et les résultats de ces morceaux ont été inclus dans le rapport.

####**5. L’option d’écho**
+ Une option connexe est echo. 

+ Lorsque l’option echo est définie sur FALSE, elle empêche le code d’apparaître dans le fichier tricoté, mais affiche les résultats du code. 

+ C’est une option que nous pouvons utiliser lors de la création de tracés. 

+ Bien que nous souhaitions que le rapport inclue le tracé, nous ne voulons peut-être pas que le rapport affiche le code qui l’a créé. 

+ Comme pour include, la valeur par défaut pour echo est également TRUE. 

+ Nous l’avons également constaté lors d’exercices précédents, car le rapport affiche à la fois le code et les résultats d’un bloc de code par défaut.

####**6. L’option d’évaluation**
+ L’option eval spécifie si nous voulons ou non évaluer le code dans un bloc de code. Si nous voulons exclure un morceau de code du rapport pour le moment, nous pouvons définir eval égal à FALSE. 

+ Lorsque eval a la valeur FALSE, le code ne s’exécute pas et les résultats ne sont pas inclus dans le rapport. 

+ Notez que le code lui-même apparaît toujours dans le rapport. 

+ Cela diffère des autres options, comme inclure, car le code n’a pas d’impact sur le rapport. 

+ Rappelez-vous, avec l’option include, même si le code ne figure pas dans le rapport de tricot, il sera toujours évalué lorsque le fichier est tricoté et aura un impact sur les segments de code suivants qui référencent des objets définis à partir du segment de code. 

+ La valeur par défaut pour eval est TRUE, car tous les segments de code sont évalués lorsque le rapport est tricoté, sauf indication contraire à l’aide de cette option.

####**7. Résumé de l’option de code**
+ Résumons chaque option dont nous avons discuté jusqu’à présent. 

+ Le code est exécuté dans le rapport lorsque les options include ou echo sont définies sur FALSE. 

+ À l’aide de l’option echo, les résultats apparaissent également dans le rapport. Lorsque eval a la valeur FALSE, le code apparaît dans le rapport, mais le code n’est pas exécuté et les résultats n’apparaissent pas.

####**8. L’option de réduction**
+ Discutons d’une autre option de morceau, l’effondrement. L’option Réduire spécifie si le code et toute sortie de texte sont divisés en plusieurs blocs ou inclus dans un seul bloc dans le rapport final. 

+ Tenez compte des messages d’avertissement que nous avons vus lors de la création de tracés. Nous pouvons modifier leur rendu dans le rapport final à l’aide de l’option de réduction. 

+ Contrairement aux autres options dont nous avons discuté jusqu’à présent, l’option par défaut pour la réduction est FAUX. 

+ Cela signifie que le code du bloc de code et la sortie de texte résultant du code seront séparés dans le rapport tricoté.

####**9. L’option de réduction**
+ Si, au lieu de cela, nous spécifions que collapse est égal à TRUE, le code et la sortie de texte sont fusionnés en un seul bloc dans le rapport tricoté.

###**Ajout d'une table des matières**

####**1. Ajout d'une table des matières**
+ Voyons comment ajouter une table des matières à un rapport.

####**2. La table des matières**
+ L'inclusion d'une table des matières donne au public un aperçu du rapport. 

+ Pour ajouter une table des matières, nous ajoutons toc et true comme nouvelle paire clé-valeur dans l'en-tête YAML. 

+ Le fichier ne sera pas tricoté si l'indentation correcte n'est pas utilisée dans l'en-tête YAML. 

+ Nous incluons le champ toc avec une indentation ajoutée, et nous l'imbriquons dans le champ **html_document**.

####**3. Profondeur de la COT**


+ Nous pouvons préciser quels en-têtes apparaîtront dans la table des matières en ajoutant le champ toc-underscore-depth. 

+ Le nombre que nous indiquons après toc_depth spécifie la taille des en-têtes qui seront listés dans la table des matières. 

+ Par exemple, une profondeur de toc de deux signifie que les en-têtes qui utilisent deux hachages, et les en-têtes plus grands qui utilisent un hachage, apparaîtront dans la table des matières. 

+ Lorsque nous tricotons le fichier, seul l'en-tête Datasets est affiché, car c'est le seul en-tête qui utilise deux hachages et il n'y a pas d'en-tête avec un seul hachage. 

+ Par défaut, la profondeur est de trois pour les documents HTML et de deux pour les documents PDF.

####**4. Numéroter les sections**


+ Nous pouvons également ajouter la numérotation des sections aux en-têtes de la table des matières en ajoutant number-underscore-sections colon true. 

+ Si nous incluons des sections numérotées et que nous voulons que la numérotation décimale commence par un, les plus grands en-têtes du rapport doivent utiliser un seul dièse.

####**5. Sections numériques**
+ Si les plus grands en-têtes du rapport commencent par deux hachages, la numérotation des sections commencera par zéro. 

+ Dans le fichier Markdown présenté, Datasets est l'en-tête le plus important et utilise deux hachages. 

+ Le numéro situé à côté de l'en-tête Datasets dans la table des matières est donc zéro-point-un-point-un.

####**6. Flottement de la table des matières**

+ Lors de la création de documents HTML, nous disposons de quelques options supplémentaires pour la table des matières. 

+ Lorsque la valeur true est attribuée à toc-underscore-float dans l'en-tête YAML, la table des matières apparaît sur le côté gauche du document et reste visible lorsque le lecteur fait défiler le document.

####**7. TOC float : collapsed**
+ Dans la marge de la table des matières, nous pouvons également utiliser les options "collapsed" et "smooth scroll".

+ Les options de table des matières précédentes que nous avons examinées, toc et toc_depth, sont des attributs distincts, tandis que les options collapsed et smooth scroll sont imbriquées dans le champ toc_float. 

+ Comme ces options sont imbriquées, nous supprimons true de toc_float lorsque nous spécifions collapsed ou smooth scroll. 

+ L'option collapsed est vraie par défaut et détermine si la table des matières affiche ou non uniquement les plus grands en-têtes. 

+ Lorsque l'option collapsed est utilisée, la table des matières se développe lorsque quelqu'un lit le rapport ou navigue vers une autre section. 

+ Lorsque la valeur "collapsed" est définie sur "false", la table des matières complète reste visible.

####**8. TOC flottante : défilement lisse**

+ L'option smooth-underscore-scroll est également vraie par défaut. 

+ Cette option anime le défilement des pages lorsqu'un lecteur clique sur un élément de la table des matières pour passer à une autre section du rapport. 

+ Lorsque l'option smooth scroll est définie sur false dans l'en-tête YAML, le fait de cliquer sur un élément de la table des matières permet au lecteur de naviguer vers cette section de l'article sans animation.

####**9. Résumé**
+ Nous avons examiné un certain nombre d'options pour l'en-tête YAML. 

+ Pour résumer, nous avons abordé les options toc, toc_depth et number_sections pour les fichiers HTML et PDF. 

+ Rappelons que la valeur par défaut de toc_depth est de trois pour les documents HTML et de deux pour les documents PDF.

+ Ensuite, nous avons parlé de toc_float pour les documents HTML, et des options collapsed et smooth_scroll dans ce champ qui peuvent être utilisées pour personnaliser davantage le rapport de tricotage.

+ Tout d’abord, nous allons modifier les styles directement dans le rapport, puis nous verrons comment créer une **feuille de style en cascade ou un fichier CSS pour ajouter les paramètres d’élément HTML à n’importe quel rapport**.

####**2. Spécification du style d’élément**


+ En dehors de tout morceau de code, nous entourons les spécifications de style avec moins de style supérieur à et inférieur à style barre oblique inverse supérieur à. 

+ Pour ajouter des éléments, nous référençons une section particulière du document avec les spécifications entourées entre accolades, en utilisant une indentation spécifique. 

+ Les propriétés couramment utilisées incluent la couleur du texte, la couleur du trait d’union d’arrière-plan, la famille de traits d’union de police et la taille du trait d’union de police. 

+ Nous pouvons inclure la section styles n’importe où dans le document, en dehors de tout bloc de code. Pour garder les sections organisées, nous allons l’ajouter en haut du document, sous l’en-tête YAML.

####**3. Style du document**


+ Commençons par le document global. 

+ Dans les indicateurs de style, nous ajoutons le corps, les accolades, la couleur, les deux-points et la couleur que nous voulons utiliser pour le texte du rapport. 

+ Nous pouvons ajouter un nom de couleur, comme rouge ou bleu, et un point-virgule. 

+ Nous n’obtiendrons pas d’erreur si nous oublions d’inclure un point-virgule après une propriété, mais le style ne sera pas mis à jour. 

+ Ici, nous spécifions rouge et voyons le texte rouge dans le fichier tricoté.

####**4. Utilisation de codes hexadécimaux couleur**


+ Nous pouvons également utiliser des codes hexadécimaux pour spécifier des couleurs, avec un seul hachage et aucun espace avant le code hexadécimal.

+ Il existe de nombreuses références que nous pouvons utiliser pour rechercher des codes hexadécimaux couleur. 

+ Ici, nous utilisons 708090 ou gris ardoise. Nous ajoutons une couleur d’arrière-plan, modifions la police de texte à l’aide de font-family, par exemple, Calibri, et tricotons le rapport.

####**5. Morceaux de code**


+ Si nous incluons des morceaux de code, nous pouvons les modifier à l’aide de précurseurs, d’accolades et en spécifiant la couleur du texte et la couleur d’arrière-plan.

####**6. La table des matières**


+ Les propriétés de la table des matières sont spécifiées à l’aide de la table des matières de hachage. 

+ Entre accolades, nous spécifions la couleur et la police du texte. 

+ Nous pouvons ajouter une couleur de bordure en utilisant la couleur du trait d’union de la bordure et modifier la taille de la police en utilisant la taille du trait d’union de police et en spécifiant un nombre suivi de px.

+ Nous ne discuterons pas de px en détail. À un niveau élevé, il correspond aux pixels et la façon dont le texte apparaît varie en fonction de l’appareil utilisé pour afficher le rapport. Nous tricotons le fichier et voyons la table des matières mise à jour.

####**7. L’en-tête**


+ Les propriétés d’en-tête sont spécifiées avec l’en-tête de hachage. 

+ Ici, nous ajoutons la couleur, la couleur d’arrière-plan, la famille de polices et la taille de police. 

+ Nous pouvons modifier la transparence de la couleur d’arrière-plan en utilisant l’opacité et une décimale. 

+ La valeur par défaut est un, et ici nous avons utilisé zéro point six. 

+ Notez que les styles d’en-tête modifient tous les aspects de l’en-tête, y compris le titre, la date et l’auteur.

####**8. Le titre, l’auteur et la date**
+ Alternativement, nous pouvons modifier chaque élément d’en-tête individuellement, en utilisant h1-dot-title pour titre, h4-dot-author pour auteur et h4-dot-date pour date.

####**9.CSS fichier**


+ Au lieu d’ajouter des styles dans le fichier, nous pouvons référencer un fichier CSS qui contient des styles spécifiques. 

+ Cela signifie que nous pouvons réutiliser un fichier CSS pour de nombreux rapports différents sans avoir besoin de copier et coller le code CSS dans chaque nouveau fichier Markdown que nous créons. 

+ Nous pouvons utiliser n’importe quel éditeur de texte pour créer la feuille de style en ajoutant les informations dans l’éditeur de texte, en excluant les mots de style de limite et en enregistrant le fichier à l’aide de l’extension point css. 

+ Ensuite, nous ajoutons un champ css à l’en-tête YAML et au nom de la feuille de style. 

+ Ici, la feuille de style est appelée styles-dot-css. Lorsque nous tricotons le fichier, les styles spécifiés apparaissent dans le rapport.

````r
---
title: "Investment Report for Projects in `r params$country`"
output: 
  html_document:
    toc: true
    toc_float: true
date: "`r format(Sys.time(), '%d %B %Y')`"
params:
  country: Brazil
  year_start: 2017-07-01
  year_end: 2018-06-30
  fy: 2018
---

<style>
body {
  color: #708090;
  font-family: Calibri;
  background-color: #F5F5F5;
}
pre {
  color: #708090;
  background-color: #F8F8FF;
}
</style>

```{r setup, include = FALSE}
knitr::opts_chunk$set(fig.align = 'center', echo = TRUE)
```

```{r data, include = FALSE}
library(readr)
library(dplyr)
library(ggplot2)

investment_annual_summary <- read_csv("https://assets.datacamp.com/production/repositories/5756/datasets/d0251f26117bbcf0ea96ac276555b9003f4f7372/investment_annual_summary.csv")
investment_services_projects <- read_csv("https://assets.datacamp.com/production/repositories/5756/datasets/bcb2e39ecbe521f4b414a21e35f7b8b5c50aec64/investment_services_projects.csv")
```


## Datasets 

### Investment Annual Summary
The `investment_annual_summary` dataset provides a summary of the dollars in millions provided to each region for each fiscal year, from 2012 to 2018.
```{r investment-annual-summary}
ggplot(investment_annual_summary, aes(x = fiscal_year, y = dollars_in_millions, color = region)) +
  geom_line() +
  labs(
    title = "Investment Annual Summary",
    x = "Fiscal Year",
    y = "Dollars in Millions"
  )
```

### Investment Projects in `r params$country`
The `investment_services_projects` dataset provides information about each investment project from 2012 to 2018. Information listed includes the project name, company name, sector, project status, and investment amounts. Projects that do not have an associated investment amount are excluded from the plot.

```{r country-investment-projects}
country_investment_projects <- investment_services_projects %>%
  filter(country == params$country) 

ggplot(country_investment_projects, aes(x = date_disclosed, y = total_investment, color = status)) +
  geom_point() +
  labs(
    title = "Investment Services Projects",
    x = "Date Disclosed",
    y = "Total IFC Investment in Dollars in Millions"
  )
```

### Investment Projects in `r params$country` in `r params$fy`
The `investment_services_projects` dataset was filtered below to focus on information about each investment project from the `r params$fy` fiscal year, and is referred to as `country_annual_investment_projects`. Projects that do not have an associated investment amount are excluded from the plot.
```{r country-annual-investment-projects}
country_annual_investment_projects <- investment_services_projects %>%
  filter(country == params$country,
         date_disclosed >= params$year_start,
         date_disclosed <= params$year_end) 

ggplot(country_annual_investment_projects, aes(x = date_disclosed, y = total_investment, color = status)) +
  geom_point() +
  labs(
    title = "Investment Services Projects",
    x = "Date Disclosed",
    y = "Total IFC Investment in Dollars in Millions"
  ) 
```
````
